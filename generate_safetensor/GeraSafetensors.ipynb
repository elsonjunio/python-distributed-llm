```python
from safetensors.torch import safe_open

shard_paths = [
    "../phi3_safetensors/model-00001-of-00004.safetensors",
    "../phi3_safetensors/model-00002-of-00004.safetensors",
    "../phi3_safetensors/model-00003-of-00004.safetensors",
    "../phi3_safetensors/model-00004-of-00004.safetensors",
]

for shard_path in shard_paths:
    with safe_open(shard_path, framework="pt") as f:
        print(f"Chaves no shard {shard_path}:")
        for key in f.keys():
            print(f" - {key}")
```
